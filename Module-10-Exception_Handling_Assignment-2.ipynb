```

# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

# ArithmeticError is a base class for all errors that occur for numeric calculations.
# It includes: ZeroDivisionError, OverflowError, FloatingPointError

# Example 1: ZeroDivisionError
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("ZeroDivisionError caught:", e)

# Example 2: OverflowError
import math

try:
    x = math.exp(1000)
except OverflowError as e:
    print("OverflowError caught:", e)
```

```

# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

# LookupError is the base class for errors raised when a key or index is not found in a collection.

# Example: KeyError
try:
    my_dict = {'a': 1}
    print(my_dict['b'])
except KeyError as e:
    print("KeyError caught:", e)

# Example: IndexError
try:
    my_list = [1, 2, 3]
    print(my_list[5])
except IndexError as e:
    print("IndexError caught:", e)
```

```

# Q5. Explain ImportError. What is ModuleNotFoundError?

# ImportError occurs when an imported module or name cannot be found.
# ModuleNotFoundError is a subclass of ImportError introduced in Python 3.6 for missing modules.

# Example: ModuleNotFoundError
try:
    import non_existing_module
except ModuleNotFoundError as e:
    print("ModuleNotFoundError caught:", e)

# Example: ImportError (trying to import a name that doesn't exist)
try:
    from math import cube_root
except ImportError as e:
    print("ImportError caught:", e)
```
